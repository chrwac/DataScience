```python
from sklearn.model_selection import train_test_split
```

```python
tds_part = titanic_dataset[["Age","Sex","Fare","Parch"]]
```

```python
X_mat = tds_part.as_matrix()
```

```python
y_mat = titanic_dataset[["Survived"]].as_matrix()
```

```python
## split into test- and training-set:
X_train,X_test,y_train,y_test = train_test_split(X_mat,y_mat,test_size=0.2,stratify=y_mat)
```

```python
## check whether both sets are actually stratified:
mean_train = np.mean(y_train)
print(mean_train)
mean_test = np.mean(y_test)
print(mean_test)
```

```python
## train a decision tree on the training set and then obtain the accuracy score on the test set:
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
## test the accuracy on the training set:
preds_train = dtc.predict(X_train)
acs_train = accuracy_score(preds_train,y_train)
print(acs_train)

## test the accuracy on the test set:
preds_test = dtc.predict(X_test)
acs_test = accuracy_score(preds_test,y_test)
print(acs_test)
```

```python
## try some alternatives first...
## logistic regression:
from sklearn.linear_model import LogisticRegression
lrc = LogisticRegression()
lrc.fit(X_train,y_train.flatten())
preds_train_lr = lrc.predict(X_train)
acs_train_lr = accuracy_score(preds_train_lr,y_train.flatten())
print(acs_train_lr)
preds_test_lr = lrc.predict(X_test)
acs_test_lr = accuracy_score(preds_test_lr,y_test.flatten())
print(acs_test_lr)
```

```python
## generalizes way better, but overall low accuracy...
```

```python
## try a random forest predictor:
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100,criterion='gini',min_samples_split=6)
rfc.fit(X_train,y_train.flatten())
preds_train_rfc = rfc.predict(X_train)
acs_train_rfc = accuracy_score(preds_train_rfc,y_train.flatten())
print(acs_train_rfc)
preds_test_rfc = rfc.predict(X_test)
acs_test_rfc = accuracy_score(preds_test_rfc,y_test.flatten())
print(acs_test_rfc)
```

```python
## try a SVM - classifier:
from sklearn.svm import SVC
svc = SVC(kernel='rbf')
svc.fit(X_train,y_train.flatten())
preds_train_svc = svc.predict(X_train)
acs_train_svc = accuracy_score(preds_train_svc,y_train.flatten())
print(acs_train_svc)
preds_test_svc = svc.predict(X_test)
acs_test_svc = accuracy_score(preds_test_svc,y_test.flatten())
print(acs_test_svc)
```

```python
from sklearn.naive_bayes import GaussianNB
nbc = GaussianNB()
nbc.fit(X_train,y_train.flatten())
preds_train_nbc = nbc.predict(X_train)
acs_train_nbc = accuracy_score(preds_train_nbc,y_train.flatten())
print(acs_train_nbc)
preds_test_nbc = nbc.predict(X_test)
acs_test_nbc = accuracy_score(preds_test_nbc,y_test.flatten())
print(acs_test_nbc)
```

```python
age_means = np.zeros(10)
surv_probs = np.zeros(10)
for i in range(0,10):
    age_range = 8
    age_min = i*age_range
    age_max = (i+1)*age_range
    age_means[i] = (age_max+age_min)/2.0
    
    num_age_range = titanic_dataset["Survived"][(titanic_dataset["Age"]>age_min)&(titanic_dataset["Age"]<=age_max)].count()
    num_surv = np.sum(titanic_dataset["Survived"][(titanic_dataset["Age"]>age_min)&(titanic_dataset["Age"]<=age_max)])
    s_prob = float(num_surv)/float(num_age_range)
    surv_probs[i] = s_prob
    print(num_age_range)
    print(s_prob)
    print("")
plt.figure(figsize=(12,9))
plt.plot(age_means,surv_probs,"o") 
plt.xlabel("age [years]")
plt.ylabel("survival probability")
```
